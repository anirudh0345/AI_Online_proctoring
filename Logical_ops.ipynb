```python
import cv2
import numpy as np 
from google.colab.patches import cv2_imshow

from google.colab import drive
drive.mount("/content/drive")
```

```python
import cv2

image1 = cv2.imread('/content/drive/MyDrive/Task_1_dataset/arith_1.jpg')
image2 = cv2.imread('/content/drive/MyDrive/Task_1_dataset/arith_2.jpg')
bitwise_and = cv2.bitwise_and(image2, image1)

cv2_imshow(bitwise_and)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
bitwise_or = cv2.bitwise_or(image2, image1)

cv2_imshow(bitwise_or)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
bitwise_xor = cv2.bitwise_xor(image2, image1)
cv2_imshow(bitwise_xor)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
bitwise_not = cv2.bitwise_not(image1)

cv2_imshow(bitwise_not)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
