```python
import cv2
import numpy as np 
from google.colab.patches import cv2_imshow

from google.colab import drive
drive.mount("/content/drive")
```

```python
image1 = cv2.imread('/content/drive/MyDrive/Task_1_dataset/arith_1.jpg')
```

```python
kernel2 = np.ones((5, 5), np.float32)/25
  
# Applying the filter
img = cv2.filter2D(src=image1, ddepth=-1, kernel=kernel2)
  
# showing the image
cv2_imshow(image1)
cv2_imshow(img)
  
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
bilateral = cv2.bilateralFilter(image1, 10, 60, 60)

cv2_imshow(bilateral)
  
cv2.waitKey()
cv2.destroyAllWindows()
```
